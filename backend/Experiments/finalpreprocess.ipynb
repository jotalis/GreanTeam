```python
import pandas as pd
```

```python
df1 = pd.read_csv('contacts_preprocessed.csv', index_col=0)
df2 = pd.read_csv('searches_preprocessed.csv', index_col=0)
```

```python
final_df = pd.merge(df1, df2, left_on='id_guest', right_on='id_user')
final_df = final_df.drop('id_user', axis=1)
```

```python
final_df.to_csv('final_preprocessed.csv')
```
