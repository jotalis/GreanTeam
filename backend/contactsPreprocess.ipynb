```python
import pandas as pd

searches_file_path = 'contacts.tsv'
searches_data = pd.read_csv(searches_file_path, sep='\t')

searches_data.head(), searches_data.dtypes

searches_data
```

```python
def mean_agg(series):
    return series.mean()

def set_agg(series):
    result = set(series.dropna().unique())
    return ', '.join(result) if result else ''

aggregation_rules = {
    'id_host': set_agg,
    'id_listing': set_agg,
    'ts_contact_at': set_agg,
    'ts_reply_at': set_agg,
    'ts_accepted_at': set_agg,
    'ts_booking_at': set_agg,
    'ds_checkin': set_agg,
    'ds_checkout': set_agg,
    'n_guests': mean_agg,
    'n_messages': mean_agg
}

# group by id_user and apply rules
grouped_searches = searches_data.groupby('id_guest').agg(aggregation_rules)

# reset the index if needed and check output
grouped_searches.reset_index(inplace=True)
print(grouped_searches.head())
```
