```python
no_outliers = pd.read_csv("no_outlier.csv")
```

```python
sns.set_style("whitegrid")
hist = sns.displot(data=no_outliers, x="n_nights",binwidth= 1,aspect=2,hue="acceptances", palette="Blues_r" )
plt.title("Number of Nights Booked to Acceptances")
plt.xlim(0,35)
plt.xlabel("Number of Nights Booked")
```

```python
hist = sns.displot(data=no_outliers, x="n_messages",binwidth= 1,aspect=2,hue="acceptances", palette="Blues_r" )
plt.title("Number of Messages to Acceptances")
plt.xlim(0,40)
plt.xlabel("Number of Messages")
```

```python
plt.figure(figsize=(20, 5))
boxplot = sns.barplot(data= no_outliers, x= "origin_country", y="n_messages", hue = "acceptances", palette="Blues_r", width=5)
plt.title("Origin Country to Number of Messages, Compares to Acceptances")
plt.xlabel("Origin Country")
plt.ylabel("Number of Messages")
```
